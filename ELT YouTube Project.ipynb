```python
#%matplotlib 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pymongo
import datetime
from flask import Flask
import pymysql
from sqlalchemy import create_engine
import json
```

```python
CA_data = "Resources/CAvideos.csv"
GB_data = "Resources/GBvideos.csv"
US_data = "Resources/USvideos.csv"

CA_data_df = pd.read_csv(CA_data)
GB_data_df = pd.read_csv(GB_data)
US_data_df = pd.read_csv(US_data)
```

```python
#CA_data_df.columns
```

```python
CA_data_df = CA_data_df.loc[:, ['video_id', 'title','trending_date','publish_time','views','likes','dislikes',
                               'comment_count','tags', 'description']]
CA_data_df.head()
```

```python
CA_data_df = CA_data_df.rename(columns={
                                      'title': 'title_ca',
                                      'trending_date': 'trending_date_ca',
                                      'publish_time': 'publish_time_ca',
                                      'views':'views_ca',
                                      'likes':'likes_ca',
                                      'dislikes':'dislikes_ca',
                                      'comment_count': 'comment_count_ca',
                                      'tags':'tags_ca',
                                    'description':'description_ca'     
                                     
                                     })

CA_data_df.columns
```

```python
GB_data_df = GB_data_df.loc[:, ['video_id', 'title','trending_date','publish_time','views','likes','dislikes',
                               'comment_count','tags', 'description']]
GB_data_df.head()
```

```python
GB_data_df = GB_data_df.rename(columns={
                                      'title': 'title_gb',
                                      'trending_date': 'trending_date_gb',
                                      'publish_time': 'publish_time_gb',
                                      'views':'views_gb',
                                      'likes':'likes_gb',
                                      'dislikes':'dislikes_gb',
                                      'comment_count': 'comment_count_gb',
                                      'tags':'tags_gb',
                                    'description':'description_gb'     
                                     
                                     })

GB_data_df.columns
```

```python
US_data_df = US_data_df.loc[:, ['video_id', 'title','trending_date','publish_time','views','likes','dislikes',
                               'comment_count','tags', 'description']]
US_data_df.head()
```

```python
US_data_df = US_data_df.rename(columns={
                                      'title': 'title_us',
                                      'trending_date': 'trending_date_us',
                                      'publish_time': 'publish_time_us',
                                      'views':'views_us',
                                      'likes':'likes_us',
                                      'dislikes':'dislikes_us',
                                      'comment_count': 'comment_count_us',
                                      'tags':'tags_us',
                                    'description':'description_us'     
                                     
                                     })

US_data_df.columns
```

```python
merged_df = pd.merge(CA_data_df, US_data_df, how ='outer', on='video_id')
merged_df.head()
```

```python
merged_df = pd.merge(merged_df, GB_data_df, how ='outer', on='video_id')
merged_df.head()
```

```python
# all_merged = pd.merge(pd.merge(CA_data_df, US_data_df, how ='outer', on='video_id'), GB_data_df, how='outer', on='video_id'))
```

```python
new_merged = merged_df.drop_duplicates(subset="video_id")
new_merged.head()
```
